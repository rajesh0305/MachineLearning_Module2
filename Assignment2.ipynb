```python
from sklearn.preprocessing import StandardScaler
standard_data = StandardScaler().fit_transform(df);

standard_data.shape
```

Implementing PCA using Scikit learn library

```python
from sklearn import  decomposition
pca = decomposition.PCA();
```

```python
pca.n_components = 2;
pca_data = pca.fit_transform(standard_data);
```

```python
pca_data = np.vstack((pca_data.T,l)).T
import seaborn as sns
#creating a new data frame which help us in plotting the result data
pca_df = pd.DataFrame(data = pca_data,columns = ("1st_principal","2nd_principal","label"))
sns.FacetGrid(pca_df,hue = "label")\
   .map(plt.scatter,'1st_principal','2nd_principal')\
   .add_legend();

plt.show()
```
